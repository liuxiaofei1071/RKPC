# 3. pandas中的None 与 NaN

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

"""pandas 中 None 与 np.nan都视作 np.nan"""
```

```python
#创建DataFrame
df = DataFrame(data=np.random.randint(0,100,size=(10,12)))
df
```

```python
#将某些数组元素赋值为nan
df.iloc[1,1] = None
df.iloc[2,4] = None
df.iloc[4,4] = None
df.iloc[2,8] = None
df.iloc[3,5] = np.nan
df
```

```python
"""将对应行的空值删除"""
'''方式1'''
# 步骤1  找到空值
df.isnull().any(axis=1)  #isnull 搭配 any 使用
# # df.notnull().all(axis=1)  #notnull 搭配 all使用 两者效果相同

# #步骤2 删除空值
drop_index = df.loc[df.isnull().any(axis=1)].index
# # drop_index = df.loc[~(df.notnull().all(axis=1))].index

# df.drop(labels=drop_index,axis=0)  #drop方式删除,axis 应用正好相反.

'''方式2'''
df.dropna()  # 默认删除行 axis = 0 ,参数可写可不写
```

### 补充axis的应用

```python
# 常规使用
axis = 1  表示行
axis = 0  表示列
# 但在drop系列中应用正好相反
df.drop(axis = 0)  #这里表示删除行
```

# pandas处理空值操作

###  dropna()  #过滤丢失数据

```python

df.dropna(axis=0)  #可以选择过滤的是行/列(默认为行),axis在drop中 0 为行,1为列
```

### 填充函数 Series/ DataFrame ---> fillna()

```python
# fillna() :value和method参数
"""method 中bfill与ffill的使用"""
df.iloc[0,1] = None
df.iloc[2,1] = None
df
```

```python
# bfill 向后填充
df.fillna(method='bfill',axis=0)
```

```python
#ffill 向前填充
df.fillna(method='ffill',axis=0)  #如下看出.还是有NaN,那是因为,前面的都是NaN,所以填充的还是NaN
```
